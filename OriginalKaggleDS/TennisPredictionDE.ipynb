## Score Transformation

Right now, the score is stored as a string and therefore does not provide any useful information. If the score is transformed into an integer, it could be used to indicate how close or one-sided a victory was.

````python
#Before the transformation give information on the string values in score column
tennis["Scorelength"] = tennis["Score"].str.len()
tennis.describe()
````

This table shows that more than 75% of the data is less than 11 strings long, so most of the matches consist of 3 or less matches.

````python
tennis20 = tennis[tennis["Scorelength"] == 19] 
tennis20.describe()
````

## Date transformation

For merging the win history and score history, it is necessary to transform the date into a numerical format or timestamp. Since the other table already stores the date as a number, we convert this table into the same format to enable the merge.

````python
print(type(tennis19.Date[0]))
tennis19["Date"] = tennis["Date"].str.replace("-","")
tennis19["Date"] = tennis19["Date"].astype(int)
````

# Win and Score History

````python
#
tennis19["Loser"] = tennis19["Player_1"]
tennis19["Loser"].loc[tennis19["Loser"] == tennis19["Winner"] ]  = tennis19["Player_2"]

tennis19 = tennis19.sort_values("Date")


tennis19["1WinB"] = (
    (tennis19["Winner"] == tennis19["Player_1"])
    .groupby(tennis19["Player_1"])
    .cumsum()
    .shift(fill_value=0)
)

tennis19["2WinB"] = (
    (tennis19["Winner"] == tennis19["Player_2"])
    .groupby(tennis19["Player_2"])
    .cumsum()
    .shift(fill_value=0)
)

tennis19["1LoseB"] = (
    (tennis19["Loser"] == tennis19["Player_1"])
    .groupby(tennis19["Player_1"])
    .cumsum()
    .shift(fill_value=0)
)

tennis19["2LoseB"] = (
    (tennis19["Loser"] == tennis19["Player_2"])
    .groupby(tennis19["Player_2"])
    .cumsum()
    .shift(fill_value=0)
)

#tennis19["1LoseB"] = 
#tennis19["2LoseB"] = 
#test = tennis19[["Player_1","Player_2","Winner","Loser"]]
#test

tennis19.describe()
````

## Last 10 Wins and Losses

````python
tennis19[["Date","Player_1", "Player_2","Loser","Winner"]]
````

````python
import pandas as pd

tennis19 = tennis19.sort_values("Date").reset_index(drop=True)
tennis19["match_id"] = tennis19.index

p1 = tennis19[["match_id", "Date", "Player_1", "Winner", "Loser"]].copy()
p1["Player"] = p1["Player_1"]
p1["Win"] = (p1["Winner"] == p1["Player"]).astype(int)
p1["Los"] = (p1["Loser"] == p1["Player"]).astype(int)

p2 = tennis19[["match_id", "Date", "Player_2", "Winner", "Loser"]].copy()
p2["Player"] = p2["Player_2"]
p2["Win"] = (p2["Winner"] == p2["Player"]).astype(int)
p2["Los"] = (p2["Loser"] == p2["Player"]).astype(int)


history = pd.concat([
    p1[["match_id", "Date", "Player", "Win", "Los"]],
    p2[["match_id", "Date", "Player", "Win", "Los"]]
])

history = history.sort_values(["Player", "Date", "match_id"])

history["WinsLast10"] = (
    history.groupby("Player")["Win"]
    .transform(lambda x: x.shift().rolling(10, min_periods=1).sum())
    .fillna(0)
)

history["LosLast10"] = (
    history.groupby("Player")["Los"]
    .transform(lambda x: x.shift().rolling(10, min_periods=1).sum())
    .fillna(0)
)
````
