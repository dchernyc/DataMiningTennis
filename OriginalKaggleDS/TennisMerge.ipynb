# Attempted Merge

Because the kaggle dataset has two advantages that the new dataset doesn't have it would be advantages to merge the two datasets.
The first advantage is that the players are already sperated into player 1 and player 2 so for the testing and training it would not be neccessary to split the new dataset.
The second advantage is that the kaggle dataset has an easy usable Score, since the new dataset has a score column with many other variables like () and ! that give more detail to the score, but make the score column not easily transformable.

The following notebook shows that the is not mergeable since the datacolumns are too diffrent.
To merge them we need on the one hand the Tournament or date of the match and the Name of both players or their Rank.
The problem with the name or rank is, that the Ranks of the 2 datasets don't match up, which indicades that the Rank is entered at diffrent times(before and after a game or before and after a tournament).
The Name column is a smaller problem since it in both entries, but both use diffrent naming conventions. So to merge them we reduce both to just the last name.
The problem to merge the entries is the tournament or date on the name. The problem with the tournament is that both datasets also use diffrent naming conventions, which makes this column unusable.
So the last column to try and merge the names is the date entry, since both date entries are now in the same format the assumption was that both of them use the same date for the matches.
However the manual looking at the unmerged table based on the column commondate shows that both datasets use anothe format again.

Since the merging is not easily possible and only around 17095 entries could be merged for an reliable result the dataset with more features should be used.


```python
import pandas as pd

csv1 = pd.read_csv("BearbTennis.csv")
csv2 = pd.read_csv("Tml_2000_2026.csv")


csv1["Date"] = csv1["Date"].astype(str)
csv2["tourney_date"] = csv2["tourney_date"].astype(str)

csv1["WinRank"] = pd.to_numeric(csv1["WinRank"], errors="coerce")
csv2["winner_rank"] = pd.to_numeric(csv2["winner_rank"], errors="coerce")

csv1["LosRank"] = pd.to_numeric(csv1["LosRank"], errors="coerce")
csv2["loser_rank"] = pd.to_numeric(csv2["loser_rank"], errors="coerce")

csv1 = csv1.dropna(subset=["WinRank", "LosRank"])
csv2 = csv2.dropna(subset=["winner_rank", "loser_rank"])

csv1["WinRank"] = csv1["WinRank"].astype(int)
csv1["LosRank"] = csv1["LosRank"].astype(int)

csv2["winner_rank"] = csv2["winner_rank"].astype(int)
csv2["loser_rank"] = csv2["loser_rank"].astype(int)

merged = pd.merge(
    csv1,
    csv2,
    left_on=["Date", "WinRank", "LosRank"],
    right_on=["tourney_date", "winner_rank", "loser_rank"],
    how="inner"
)
```
